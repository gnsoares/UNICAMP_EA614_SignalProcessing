```python
import numpy as np
from numpy import abs, linspace, pi, sin
from numpy.fft import fft, ifft
from scipy.signal import freqz
import matplotlib.pyplot as plt
```

```python
x = np.zeros(6)
for i in range(x.size):
    x[i] = 6 - i
h = np.zeros(6)
h[0] = h[1] = 1
h[2] = -1
```

```python
y = np.array(
    [
        np.sum(np.roll(h[::-1], i + 1)*x) for i in range(6)
    ]
)
```

```python
y
```

```python
for i in range(1, 7):
    print(np.roll(h[::-1], i), x, np.roll(h[::-1], i)*x)
    print()
```

```python
np.convolve(x, h[:3])
```
