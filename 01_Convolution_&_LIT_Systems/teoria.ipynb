```python
# Bibliotecas utilizadas
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 75
k = 50
d = 20
n = np.arange(-i, i+1, 1)
loc0 = np.where(n == 0)[0][0]
s = np.zeros(n.size)
s[loc0:loc0+k] = 1.5*np.ones(k)
h = np.zeros(n.size)
h[loc0:loc0+d] = np.ones(d)

nx = [0, 10, 34, 50, 60 , 69]

labels = ['', 'D/2', 'n', 'K', 'K+D/2', 'K+D-1']
hlabel = {k: r'$h[{}-k]$'.format(v) for k, v in zip(nx, labels)}

s[s==0] = np.nan
h[h==0] = np.nan
```

```python
fig, axs = plt.subplots(len(nx), 1, figsize=(15, 15))
plt.rcParams.update({'font.size': 18})
beg = np.where(n == -i+3*d/2)[0][0]
for ax, label, j in zip(axs, labels, nx):
    ax.stem(n[beg:], s[beg:], basefmt='k', label=r'$s[k]$')
    markerline, stemlines, baseline = ax.stem(n[beg:], np.roll(h, -j)[::-1][beg:], basefmt='k', markerfmt='o', label=hlabel[j])
    markerline.set_markerfacecolor('orange')
    markerline.set_markeredgecolor('orange')
    for stemline in stemlines:
        stemline.set_color('orange')
    ax.text(85, .75, r'$n = {}$'.format(label or 0) if label != 'n' else r'$\forall n$', color="k")
    ax.legend(loc=2)
    ax.set_xticks([-d, 0, d, k, k+d])
    ax.set_xticklabels([r'$-D$', 0, r'$D$', r'$K$', r'$K+D$'])
    ax.set_yticks([1, 1.5])
    ax.set_yticklabels([r'$y_h$', r'$y_s$'])
plt.show()
```

```python
fig.tight_layout()
fig.savefig('a.png')
```
